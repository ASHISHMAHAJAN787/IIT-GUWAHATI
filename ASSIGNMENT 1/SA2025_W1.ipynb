### 1. What is name of car that has the highest horsepower?  

```python
max_hp = df['horsepower'].max()
max_hp_car = df[df['horsepower'] == max_hp]['name'].iloc[0]

print(f"\nThe car with highest horsepower ({max_hp}) is: {max_hp_car}")
```

### 2. How many cars have mpg  35?

```python
high_mpg_cars = df[df['mpg'] >= 35]

num_high_mpg_cars = len(high_mpg_cars)

print(f"Number of cars with MPG  35: {num_high_mpg_cars}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

origin_counts = filtered_cars['origin'].value_counts()

most_common_origin = origin_counts.idxmax()
most_common_origin_count = origin_counts.max()

print(f"Most common origin: {most_common_origin} (appears {most_common_origin_count} times)")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japanese_cars = df[df['origin'].str.lower().str.contains('japan')]
mean_acceleration = japanese_cars['acceleration'].mean().round(2)

print(mean_acceleration)
```

### 5. Which year had the highest average mpg?

```python
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

best_year = avg_mpg_by_year.idxmax()
highest_avg_mpg = round(avg_mpg_by_year.max(), 2)  

print(f"The year with highest average MPG is {best_year} with {highest_avg_mpg} MPG")

```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
df = df.reset_index()

median_mpg = df['mpg'].median()

high_mpg_cars = df[df['mpg'] > median_mpg]

max_ratio = high_mpg_cars['horsepower_to_weight'].max()

best_cars = high_mpg_cars[high_mpg_cars['horsepower_to_weight'] == max_ratio]

print(f"Median MPG threshold: {median_mpg:.1f}")
print(f"Best hp/weight ratio among efficient cars: {max_ratio:.6f}\n")
print("Top performing car(s):")
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'horsepower_to_weight']].to_string(index=False))
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

avg_mpg['origin'] = avg_mpg['origin'].str.capitalize()

plt.figure(figsize=(12, 7))
for origin in ['Usa', 'Japan', 'Europe']:
    origin_data = avg_mpg[avg_mpg['origin'] == origin]
    plt.plot(origin_data['model_year'],
             origin_data['mpg'],
             label=origin,
             marker='o',
             linewidth=3)

plt.title('Average MPG Trends by Country of Origin', fontsize=14)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.xticks(avg_mpg['model_year'].unique())
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
palette = {
    'japan': '#e41a1c',  
    'europe': '#377eb8',  
    'usa': '#4daf4a'     
}

plt.figure(figsize=(12, 7))

scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    palette=palette,  
    sizes=(30, 200),
    alpha=0.8,
    edgecolor='w',
    linewidth=0.5
)

plt.title("Vehicle Weight vs Horsepower\n(Size = MPG, Color = Origin)", fontsize=14, pad=20)
plt.xlabel("Horsepower (hp)", fontsize=12)
plt.ylabel("Weight (lbs)", fontsize=12)

legend = plt.legend(
    title='Origin / MPG',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    frameon=True,
    framealpha=1
)

for handle in legend.legend_handles:
    handle.set_alpha(1)  

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
g = sns.PairGrid(df, vars=['horsepower', 'weight', 'mpg'], hue='origin', hue_order=['japan', 'europe', 'usa'])
g.map_upper(sns.scatterplot, size=df['mpg'], alpha=0.7)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde=True)
g.add_legend(title='Origin')
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
multi_year_cars = df.groupby('name').filter(lambda x: len(x) > 1)

consistency_stats = (multi_year_cars.groupby('name')['mpg']
                     .agg(['count', 'std', 'mean'])
                     .reset_index()
                     .rename(columns={'count': 'appearances', 'std': 'mpg_std', 'mean': 'avg_mpg'}))

consistent_cars = consistency_stats[consistency_stats['mpg_std'] < 1.0]

result = (consistent_cars.sort_values(['appearances', 'avg_mpg'], 
                                    ascending=[False, False])
          .reset_index(drop=True))

print("Most Consistent Car Models (by MPG):")
print(result.to_string(index=False))

for model in result['name'].head(3):
    model_data = df[df['name'] == model][['model_year', 'mpg']].sort_values('model_year')
    print(f"\n{model} MPG by Year:")
    print(model_data.to_string(index=False))
```
